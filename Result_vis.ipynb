```python
import torch
import torch.nn as nn
import torch
import pandas as pd
import numpy as np
from typing import Union, List, Dict, Any, cast
import matplotlib.pyplot as plt
```

```python
def calcualte(dirc):
    spa = np.load(dirc)
    X,Y = np.meshgrid(range(64), range(64))
    spa = spa.reshape((64,64))
    c_spa = np.zeros((64,64))
    c_spa[spa>0.85] = 1
    c_spa[spa<-0.85] = -1
    return X,Y,c_spa
    
```

```python
from matplotlib.ticker import FuncFormatter
from matplotlib import ticker
def purity(dirc):
    spa = np.load(dirc)
    spa = spa.reshape((64,64))
    c_spa = np.zeros((64,64))
    c_spa[spa>0.85] = 1
    c_spa[spa<-0.85] = -1
    position = [0,8,14,20,26,32,38,44,50,56,64]
    area = np.zeros((10,10))
    face = np.zeros((10,10))
    for x in range(10):
        for y in range(10):
            number = c_spa[position[x]:position[x+1],position[y]:position[y+1]]
            area[x,y] = (position[x+1] -position[x])*(position[y+1] -position[y])
            face[x,y] = len(number[number>0])
    percent = face/area
    (r,c) = np.unravel_index(np.argmax(percent), percent.shape)
    
    area_loc = {}
    face_loc = {}
    for x in range(10):
        for y in range(10):
            x_r = abs(x-r)
            y_c = abs(y-c)
#             if abs(x-r)>5:
#                 x_r = abs(x-r-5)
#             else:
#                 x_r = abs(x-r)
                
#             if abs(y-c)>5:
#                 y_c = abs(y-c-5)
#             else:
#                 y_c = abs(y-c)
            loc = x_r**2 + y_c**2
            face_loc[loc] = []
            area_loc[loc] = []

    for x in range(10):
        for y in range(10):
            x_r = abs(x-r)
            y_c = abs(y-c)
#             if abs(x-r)>5:
#                 x_r = abs(x-r-5)
#             else:
#                 x_r = abs(x-r)

#             if abs(y-c)>5:
#                 y_c = abs(y-c-5)
#             else:
#                 y_c = abs(y-c)
            loc = x_r**2 + y_c**2
            face_loc[loc].append(face[x,y])
            area_loc[loc].append(area[x,y])
    final = {}   

    for k in sorted(list(face_loc.keys())):
        final[k] = sum(face_loc[k])/sum(area_loc[k])
        
#     plt.plot(np.array(np.sqrt(list(final.keys())))[:20],np.array(list(final.values()))[:20])
    return np.array(np.sqrt(list(final.keys())))[:20], np.array(list(final.values()))[:20]


    plt.plot(x,np.mean(y,axis=0))
    plt.fill_between(x,y1=np.mean(y,axis=0)+ np.std(y,axis=0),y2=np.mean(y,axis=0)-np.std(y,axis=0)
                     ,color='#3bb300',alpha=0.2)
    

            
            
```

# Response Profile

```python
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
def func(x, a, b,c):
    return a * np.exp(-b * x)+c 
# def func(x, A, B, C, sigma):

#     return A*np.exp(-(x-B)**2/(2*sigma**2)) + C
```

```python
corr = np.load('/Users/hernando/Desktop/thesis/correlation2.npy',allow_pickle=True).item()
```

```python
x = []
y = []
absy=[]
absx=[]
for i in corr.keys():
    x.extend([i]*len(corr[i]))
    y.extend(corr[i])
for i in range(len(y)):
    if y[i]>=0:
        absy.append(y[i])
        absx.append(x[i])
        
```

```python
plt.figure(figsize=(6,4))
plt.scatter(x,y,color='black',s=5)
popt, pcov = curve_fit(func, absx, absy)
plt.plot(np.arange(0,13,0.5), func(np.arange(0,13,0.5), *popt), 'r-',)
plt.xlabel('Pair-wise Cortical Distance $($mm$)$',size=18)
plt.ylabel('Pair-wise Response Correlation',size=18)
plt.title('TVAE-FC7',size=15)
plt.savefig('./plot/RC_tvaefc7.png',dpi=200,bbox_inches='tight')
```
